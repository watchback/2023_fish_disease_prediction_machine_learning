```python
from scipy.stats import mannwhitneyu
```

```python
feature_list = []
for i in range(len(data.columns)-1):
    print(data.columns[i])
    udata = data.iloc[:,[i,-1]]
    jeong = udata[udata['결과'] == 0]
    jeong = jeong.drop(['결과'],axis=1)
    jooE = udata[udata['결과'] == 1]
    jooE = jooE.drop(['결과'],axis=1)
    zeel = udata[udata['결과'] == 2]
    zeel = zeel.drop(['결과'],axis=1)
    
    U1, p = mannwhitneyu(jeong, jooE, alternative='two-sided')
    U2, p2 = mannwhitneyu(jooE, zeel, alternative='two-sided')
    U3, p3 = mannwhitneyu(zeel, jeong, alternative='two-sided')
    print(p,p2,p3)
    if p<(0.05/309) and p2<(0.05/309) and p3<(0.05/309): # (80개의 샘플 - 전부 0인 특성 4개) * 샘플 수 3
        feature_list.append(udata.columns[0])
```
