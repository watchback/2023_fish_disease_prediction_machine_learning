```python
import sys
import pandas as pd
import pickle
import numpy as np
import warnings

warnings.filterwarnings(action='ignore')
with open('/Users/csb/Desktop/Project/넙치 머신러닝/양식어류 질병예측 모델 최종본/10featureModel.pkl', 'rb') as file:
    model = pickle.load(file)
```

```python
data = pd.read_csv('/Users/csb/Desktop/Project/넙치 머신러닝/양식어류 질병예측 모델 최종본/10featureSampledata.txt', sep="\t")
```

```python
selectedFeatures_ofStat = ["ID","Temperature", "Condition factor", "T P", "Wgp2", "Wgi3", "Whkp2",
                           "Whki2-1", "Wlr1", "Wlr2-1", "Wsi2"]
```

```python
data = data.loc[:,selectedFeatures_ofStat]
data = pd.DataFrame.dropna(data, axis=0)
```

```python
feature_columns = list(data.columns.difference(['ID']))
idCol = data.loc[:,"ID"]
test_x = data[feature_columns]
```

```python
result_df = pd.DataFrame(columns=['ID', 'Predicted_CLASS', 'Max_Value'])
mapping_dict = {1 : 'Normal', 2 : 'Warning', 3 : 'Infection'}
n_rows, n_cols = data.shape
for i in range(n_rows):
    test_data = pd.DataFrame(test_x.iloc[i].values.reshape(1, -1))
    predicted_prob = model.predict_proba(test_data)[0]
    max_value = np.max(predicted_prob)
    predicted_class = model.predict(test_data)[0]
    result_df.loc[i] = [idCol[i], predicted_class, max_value ]
result_df['Predicted_CLASS'] = result_df['Predicted_CLASS'].map(mapping_dict)
result_df['Predicted_CLASS'] = result_df['Predicted_CLASS'] + ' ( ' + round(result_df['Max_Value']*100,1).map(str) + '% )'
result_df = result_df.drop('Max_Value', axis=1)
```
